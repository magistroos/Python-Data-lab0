```python
def gnome_sort(arr):
    index = 0
    while index < len(arr):
        if index == 0:
            index += 1
        if arr[index] >= arr[index - 1]:  # Порівнюємо правильним чином
            index += 1
        else:
            arr[index], arr[index - 1] = arr[index - 1], arr[index]
            index -= 1
    return arr

def task_8():
    while True:
        user_input = input("Enter a list of numbers separated by spaces: ")
        try:
            # Конвертуємо вхідну строку у список цілих чисел
            input_list = list(map(int, user_input.split()))
            sorted_list = gnome_sort(input_list)
            print("Sorted list:", sorted_list)
            break  # Вихід з циклу після успішного сортування
        except ValueError:
            print("Invalid input. Please enter only numbers separated by spaces.")

task_8()
```
